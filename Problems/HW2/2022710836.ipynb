```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import *
from matplotlib import cm

X1 = np.genfromtxt('./hw2_data-1/hw2_2_X_1.csv', delimiter=',')
X2 = np.genfromtxt('./hw2_data-1/hw2_2_X_2.csv', delimiter=',')
X3 = np.genfromtxt('./hw2_data-1/hw2_2_X_3.csv', delimiter=',')

y1 = np.genfromtxt('./hw2_data-1/hw2_2_y_1.csv', delimiter=',')
y2 = np.genfromtxt('./hw2_data-1/hw2_2_y_2.csv', delimiter=',')
y3 = np.genfromtxt('./hw2_data-1/hw2_2_y_3.csv', delimiter=',')
```

```python
def problem_a(X, y):
    clf = SVC(kernel="linear")
    clf.fit(X,y)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.scatter(clf.support_vectors_[:, 0],clf.support_vectors_[:, 1], s=100,facecolors="none", edgecolors="black")
    xlim = (np.min(X.T[0])-0.7, np.max(X.T[0])+0.7)
    ylim = (np.min(X.T[1])-0.7, np.max(X.T[1])+0.7)
    xx = np.linspace(xlim[0], xlim[1], 100)
    yy = np.linspace(ylim[0], ylim[1], 100)
    XX, YY = np.meshgrid(xx, yy)
    plt.xlabel("X1")
    plt.ylabel("y1")
    g = clf.decision_function(np.vstack([XX.ravel(), YY.ravel()]).T).reshape(XX.shape)

    plt.contour(XX, YY, g, colors='k',levels=[-1, 0, 1], linestyles=['--', '-', '--'])
    
```

```python
problem_a(X1, y1)
```

```python
def problem_b(X, y):
    clf = SVC(kernel="rbf")
    clf.fit(X,y)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.scatter(clf.support_vectors_[:, 0],clf.support_vectors_[:, 1], s=100,facecolors="none", edgecolors="black")
    xlim = (np.min(X.T[0])-0.1, np.max(X.T[0])+0.1)
    ylim = (np.min(X.T[1])-0.1, np.max(X.T[1])+0.1)
    xx = np.linspace(xlim[0], xlim[1], 100)
    yy = np.linspace(ylim[0], ylim[1], 100)
    XX, YY = np.meshgrid(xx, yy)
    plt.xlabel("X2")
    plt.ylabel("y2")
    g = clf.decision_function(np.vstack([XX.ravel(), YY.ravel()]).T).reshape(XX.shape)

    plt.contour(XX, YY, g, colors='k',levels=[-1, 0, 1], linestyles=['--', '-', '--'])
```

```python
problem_b(X2, y2)
```

```python
def problem_c(X, y, c):
    clf = SVC(kernel="linear", C=c)
    clf.fit(X,y)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.scatter(clf.support_vectors_[:, 0],clf.support_vectors_[:, 1], s=100,facecolors="none", edgecolors="black")
    xlim = (np.min(X.T[0])-0.1, np.max(X.T[0])+0.1)
    ylim = (np.min(X.T[1])-0.1, np.max(X.T[1])+0.1)
    xx = np.linspace(xlim[0], xlim[1], 100)
    yy = np.linspace(ylim[0], ylim[1], 100)
    XX, YY = np.meshgrid(xx, yy)
    plt.xlabel("X3")
    plt.ylabel("y3")
    g = clf.decision_function(np.vstack([XX.ravel(), YY.ravel()]).T).reshape(XX.shape)

    plt.contour(XX, YY, g, colors='k',levels=[-1, 0, 1], linestyles=['--', '-', '--'])
```

```python
#C is 0.001
problem_c(X3, y3, 0.001)
```
